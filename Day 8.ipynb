```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from scipy.stats import boxcox
date = pd.date_range(start='2020-01-01', periods=60, freq='M')
sales = np.random.randint(200,500,60)

df = pd.DataFrame({
    "Date":date,
    "Sales":sales
})

df.set_index("Date", inplace=True)

print(df.head())
dataset = df.copy()
print(dataset.describe())
dataset.rename(columns={"Sales":"y"}, inplace=True)
print(dataset.head())
plt.figure(figsize=(10,5))
plt.plot(dataset)
plt.title("Time Series Data")
plt.show()
decomposition = seasonal_decompose(dataset, model='additive')

decomposition.plot()

plt.show()
dataset["sin"] = np.sin(np.arange(len(dataset)))
dataset["cos"] = np.cos(np.arange(len(dataset)))

dataset["boxcox"], lam = boxcox(dataset["y"])

print(lam)
model = ExponentialSmoothing(dataset["y"], trend="add", seasonal=None)

fit = model.fit()

forecast = fit.forecast(10)

print(forecast)
hw_model = ExponentialSmoothing(
    dataset["y"],
    trend="add",
    seasonal="add",
    seasonal_periods=12
)

hw_fit = hw_model.fit()

hw_forecast = hw_fit.forecast(10)

print(hw_forecast)
model = ARIMA(dataset["y"], order=(1,1,1))

fit = model.fit()

print(fit.summary())
model = SARIMAX(dataset["y"],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

fit = model.fit()

print(fit.summary())
model = SARIMAX(dataset["y"],
                exog=dataset[["sin","cos"]],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

fit = model.fit()

print(fit.summary())
print("AIC:", fit.aic)
print("BIC:", fit.bic)
```
