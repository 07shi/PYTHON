```python
import pandas as pd
data = {'Marks': [35, 67, 89, 45, 72, 90, 55, 60]}
df = pd.DataFrame(data)
df['Equal_Freq_Bins'] = pd.qcut(df['Marks'], q=3)
print(df)

```
